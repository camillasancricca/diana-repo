```python
import pandas as pd
import numpy as np
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
rankings = pd.read_csv("../results/techniques_completeness_evaluation_total.csv")
rankings
```

```python
dd = rankings.ranking_real
```

```python
def get_ranking_array(s):
    s = s.replace(".        ]",".0]")
    s = s.replace("   "," ")
    s = s.replace("  "," ")
    s = s.replace(" ]","]")
    s = s.replace("\'","")
    s = s.replace("[","")
    s = s.replace("]","")
    s = s.split(" ")
    s = np.array(s)
    return s
```

```python
res = []
for i in range(0,len(dd)):
    res.append(get_ranking_array(dd[i]))
```

```python
df = pd.DataFrame(res)
```

```python
df = pd.read_csv("../results/rankings_imputation_real.csv")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
max_count = df.apply(pd.Series.value_counts).max().max()
fig, axes = plt.subplots(nrows=1, ncols=len(df.columns), figsize=(20, 5), sharey=True)
colors = sns.color_palette("hsv", len(df.columns))

for i, column in enumerate(df.columns):
    sns.countplot(y=column, data=df, ax=axes[i], palette=colors)
    axes[i].set_title(column)
    axes[i].set_xlabel('Frequency')
    axes[i].set_ylabel('')
    axes[i].set_xlim(0, max_count)  # Impostiamo il limite dell'asse x

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

melted_df = df.melt(var_name='Position', value_name='Technique')

melted_df['Position'] = melted_df['Position'].apply(lambda x: df.columns.get_loc(x) + 1)

plt.figure(figsize=(15, 8))
sns.boxplot(x='Technique', y='Position', data=melted_df, order=sorted(melted_df.Technique.unique()))
plt.xticks(rotation=90)
plt.title('Distribution of the positions of the techniques')
plt.xlabel('Data Imputation')
plt.ylabel('Ranking Position')
plt.gca().invert_yaxis()
plt.show()
```
